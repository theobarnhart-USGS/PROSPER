```python
import gdal
from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset
```

```python
fl = './N_America_model_wtd_v2.nc'
```

```python
ds = Dataset(fl)
```

```python
lats = ds.variables['lat']
lons = ds.variables['lon']
```

```python
lon_grid,lat_grid = np.meshgrid(lons,lats)
```

```python
wtd = ds.variables['WTD'][0]
```

```python
m = Basemap(llcrnrlon = -125, urcrnrlon = -115, llcrnrlat = 40, urcrnrlat = 47, projection='laea',lat_0=45,lon_0=-124)
```

```python
plt.figure(figsize=(10,10))
xmin = 5500
xmax = 8000
ymin = 4000
ymax = 5500
x,y = m(lon_grid[ymin:ymax,xmin:xmax],lat_grid[ymin:ymax,xmin:xmax])
m.pcolormesh(x,y,wtd[ymin:ymax,xmin:xmax],edgecolor=None, rasterized=True,vmin=0,vmax=200,cmap='YlGn_r')
m.drawstates(linewidth=2)
plt.colorbar(label='Depth to Groundwater [m]')
```

```python
plt.figure(figsize=(10,10))
plt.pcolormesh(wtd[4500:5000,6500:8000],rasterized=True,edgecolors=None,cmap='YlGn_r',vmax=200)
plt.axis('equal')
plt.colorbar()
```
